```python
# Count plotting in logarithmic scale

def uniplot(df,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='magma') 
        
    plt.show()
```

```python
plt.figure(figsize = (15,7))

corr_matrix = num_df.corr()
sns.heatmap(corr_matrix, annot =True)
```

```python
corrdf = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
```

```python
corrdf = corrdf.unstack().reset_index()
```

```python
corrdf.columns = ['Var1', 'Var2', 'Correlation']
corrdf
```

```python
corrdf.dropna(subset = ['Correlation'], inplace = True)
```

```python
# corrdf.Correlation = abs(corrdf.Correlation)
# corrdf.sort_values('Correlation', ascending = False)
```

```python
positive = corrdf[corrdf['Correlation'] > 0].sort_values(by='Correlation', ascending)
negative = corrdf[corrdf['Correlation'] < 0].sort_values(by='Correlation')
```

```python
# Finding some correlation for numerical columns for both target 0 and 1 

target0_corr=df0.iloc[0:,2:]
target1_corr=df1.iloc[0:,2:]

target0=target0_corr.corr(method='spearman')
target1=target1_corr.corr(method='spearman')
```

```python
# Correlation for target 1

target1
```

```python
# Now, plotting the above correlation with heat map as it is the best choice to visulaize

# figure size

def targets_corr(data,title):
    plt.figure(figsize=(15, 10))
    plt.rcParams['axes.titlesize'] = 25
    plt.rcParams['axes.titlepad'] = 70

# heatmap with a color map of choice


    sns.heatmap(data, cmap="RdYlGn",annot=False)

    plt.title(title)
    plt.yticks(rotation=0)
    plt.show()
```

```python
# For Target 0

targets_corr(data=target0,title='Correlation for target 0')
```

```python

# For Target 1

targets_corr(data=target1,title='Correlation for target 1')
```
