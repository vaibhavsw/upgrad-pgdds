```python
import numpy as np 
import pandas as pd

# Defining the three dataframes indicating the gold, silver, and bronze medal counts
# of different countries
gold = pd.DataFrame({'Country': ['USA', 'France', 'Russia'],
                         'Medals': [15, 13, 9]}
                    )
silver = pd.DataFrame({'Country': ['USA', 'Germany', 'Russia'],
                        'Medals': [29, 20, 16]}
                    )
bronze = pd.DataFrame({'Country': ['France', 'USA', 'UK'],
                        'Medals': [40, 28, 27]}
                    )

df_1 = pd.concat([gold, silver, bronze])
df_1['Medals'] = df_1['Medals'].astype(float)
df_1 = df_1.fillna(0.0).groupby('Country').sum(axis=1).sort_values('Medals', ascending=False)

print(df_1)
```
