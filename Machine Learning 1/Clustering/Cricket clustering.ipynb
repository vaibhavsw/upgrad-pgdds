```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import datetime as dt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
data = pd.read_csv('./Cricket.csv',sep=",", encoding="ISO-8859-1", header=0)
data.head()

data['HS'] = data.HS.apply(lambda x: x.replace('*', ''))

data.head()
```

```python

num_data = data.drop(['Player', 'Span'], axis=1)

scaler = StandardScaler()
data_scaled = scaler.fit_transform(num_data)
data_scaled
```

```python
rfm_df_scaled = pd.DataFrame(data_scaled, columns=[num_data.columns])
rfm_df_scaled.head()
```

```python
kmeans = KMeans(n_clusters=4, random_state=100)
kmeans.fit(rfm_df_scaled)
```

```python
kmeans.labels_
```

```python
rfm_df_scaled['cluster'] = kmeans.labels_
rfm_df_scaled.head()
```

```python
rfm_df_scaled['player'] = data.Player
```
